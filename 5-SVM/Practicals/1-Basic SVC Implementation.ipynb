## Support Vector Classifier Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Lets create synthetic data points
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(
    n_samples=1000,
    n_features=2,
    n_classes=2,
    n_clusters_per_class=2,
    n_redundant=0
)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0] # Converting into dataframe and getting first column
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
svc.coef_
```

```python
## Prediction
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print("Confusion Matrix :- \n",confusion_matrix(y_test,y_pred))
```

### Trying with RBF kernal

```python
rbf=SVC(kernel='rbf')
```

```python
rbf.fit(X_train,y_train)
```

```python
## Prediction
y_pred1=rbf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred1))
print("Confusion Matrix :- \n",confusion_matrix(y_test,y_pred1))
```

### Trying with Polnomial Kernal

```python
polynomial=SVC(kernel='poly')
polynomial.fit(X_train,y_train)
## Prediction
y_pred2=polynomial.predict(X_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

### Trying with Sigmoid kernal

```python
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
## Prediction
y_pred3=sigmoid.predict(X_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

```python
sigmoid.intercept_
```

## Hyperparametere Tuning With SVC

```python
from sklearn.model_selection import GridSearchCV
 
# defining parameter range
param_grid = {
            'C': [0.1, 1, 10, 100, 1000],
            'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
            'kernel': ['rbf'] # For hyperparameter tunning selected 'rbf', bcoz it's higher accuracy
}
```

```python
grid=GridSearchCV(SVC(), 
                  param_grid=param_grid, 
                  refit=True, # Refit an estimator using the best found parameters on the whole dataset.
                  cv=5, 
                  verbose=3 # controls the amount of logging output during the fitting process
                )
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
## Prediction
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
